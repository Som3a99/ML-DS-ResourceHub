```python
data = pd.read_csv("Global Missing Migrants Dataset.csv")
cnt = 0
for c in data.columns:
    cnt += len(data[data[c].isnull()])
print(cnt/len(data))

# Fill data with mean / mode
for c in data.columns:
    if data[c].isnull().any():
        # print(c)
        if(data[c].dtype == 'O'):
            data[c] = data[c].fillna(data[c].mode().values[0])
        else:
            data[c] = data[c].fillna(data[c].mean())

cnt  = 0          
for c in data.columns:
    cnt += len(data[data[c].isnull()])
print(cnt/len(data))

# Encoding using one hot encoder
from sklearn.preprocessing import OneHotEncoder

for c in data.columns:
    if data[c].dtype == 'O':
        oe = OneHotEncoder()
        tmp = pd.DataFrame(oe.fit_transform(data[[c]]).todense(),columns = oe.get_feature_names_out())
        # print(tmp)
        data =  pd.concat([data,tmp],axis=1)
        data = data.drop([c],axis=1)
        # break
print(data.columns)
```

```python
data = pd.read_csv("Global Missing Migrants Dataset.csv")
cnt = 0
for c in data.columns:
    cnt += len(data[data[c].isnull()])
print(cnt/len(data))

# Fill data with mean / mode
for c in data.columns:
    if data[c].isnull().any():
        # print(c)
        if(data[c].dtype == 'O'):
            data[c] = data[c].fillna(data[c].mode().values[0])
        else:
            data[c] = data[c].fillna(data[c].mean())

cnt  = 0          
for c in data.columns:
    cnt += len(data[data[c].isnull()])
print(cnt/len(data))

# Encoding using one hot encoder
from sklearn.preprocessing import LabelEncoder

for c in data.columns:
    if data[c].dtype == 'O':
        le = LabelEncoder()
        data[c] = le.fit_transform(data[c])
        # break
print(data.columns)
```
